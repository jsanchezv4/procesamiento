# üîé Introducci√≥n r√°pida a Expresiones Regulares (Regex)

Las expresiones regulares (o regex) son una forma poderosa de buscar, extraer, validar y limpiar texto usando patrones.


En NLP se usan todo el tiempo para:
- Limpiar texto antes de vectorizar
- Extraer informaci√≥n espec√≠fica (emails, fechas, n√∫meros, hashtags, etc.)
- Normalizar textos
- Detectar patrones ling√º√≠sticos simples

En Python trabajamos con regex usando el m√≥dulo est√°ndar `re`.

````python
import re
````

````python
texto = "Escr√≠beme a miguel@example.com o a contacto@empresa.co"
````

````python
patron = r"\S+@\S+\.\S+"
re.search(patron, texto)

#Si encuentra el patr√≥n, devuelve un objeto Match; si no, devuelve None.
````

**Lectura de la regla**
- Una o m√°s cosas que no sean espacios
- seguidas de @
- seguidas de una o m√°s cosas que no sean espacios
- seguidas de .
- seguidas de una o m√°s cosas que no sean espacios

¬øC√≥mo consigo todas las coincidencias?

````python
re.findall(patron, texto)
````

### üßπ Limpieza t√≠pica en NLP

**C√≥mo se lee paso a paso**
- [...] ‚Üí conjunto de caracteres
- ^ (dentro de []) ‚Üí negaci√≥n ‚Üí ‚Äútodo lo que NO sea‚Ä¶‚Äù
- a-z ‚Üí letras min√∫sculas
- A-Z ‚Üí letras may√∫sculas
- √°√©√≠√≥√∫√±√Å√â√ç√ì√ö√ë ‚Üí letras con tilde y √±
- \s ‚Üí espacios (espacio, tab, salto de l√≠nea)
